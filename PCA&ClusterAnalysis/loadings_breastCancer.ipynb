#Loadings
Loadiings of the PCA determines how the variables contribute to each of the principal components. 

There are two types of loadings.

<li>Positive Loading

<li>Negative Loading

Positive Loadings = > Positive Correlation. i.e an increase in one results an increase in other.

Negative Loadings = > Negative Correlation. i.e a decrease in one results in an increase in other. 

Large value of loadings indicate that the variable has strong effect on the principal components.

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

X = cancer.data
Y = cancer.target

df =pd.DataFrame(X,columns = cancer.feature_names)
df.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
x_pca = pca.fit_transform(X)
x_pca
```

```python
loadings = pd.DataFrame(pca.components_.T, columns = ['PC1','PC2'], index = cancer.feature_names)
loadings
```

```python
loadings = pca.components_.T * np.sqrt(pca.explained_variance_ratio_)

loadings_matrix = pd.DataFrame(loadings, columns = ['PC1','PC2'], index = cancer.feature_names)
loadings_matrix
```
