# SCREE Plot

SCREE plot is used to visualize the percentage of variation explained by each principal component.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#import the dataset
url = ("https://raw.githubusercontent.com/JWarmenhoven/ISLR-python/master/Notebooks/Data/USArrests.csv")

data = pd.read_csv(url)
data.head()
```

```python
#We need only columsn specified as crimes for PCA
# so extract those columns
df = data.iloc[:,1:5]
df.head()
```

```python
#Perform PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
pca = PCA(n_components = 4)
x_pca = pca.fit_transform(scaler.fit_transform(df))
x_pca
```

```python
#Sketch the SCREE Plot
PC_values = np.arange(pca.n_components_)+1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth = 2, color = 'g')
plt.show()
```
